# 動画のダウンロード
resultフォルダーにある動画を全て自動でダウンロードします。手動でダウンロードする場合は、スキップして下さい。

```
from google.colab import files
for filename in os.listdir("./results"):
  extension = os.path.splitext(filename)[-1].lower()
  if extension == ".mp4":
    files.download("./results/" + filename)
```
